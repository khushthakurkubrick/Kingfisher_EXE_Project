```python
import requests 
import json
from bs4 import BeautifulSoup
```

```python
url = r'https://autochek.africa/ng/cars-for-sale?country=ng&page_size=100&transmission=automatic&page_number=1'
r = requests.get(url)
```

```python
import pandas as pd 
soup = BeautifulSoup(r.text, 'html.parser')

tag = 'script'
attributes = {'id':'__NEXT_DATA__'}
table_soup = soup.find(tag, attributes)

for item in table_soup:
    x = json.loads(item)
    y = x['props']['pageProps']['cars']
    df = pd.DataFrame(y)
    
n_pages = 54
url = 'https://autochek.africa/ng/cars-for-sale?country=ng&page_size=100&transmission=automatic&page_number='

link = []

for i in range (1,n_pages+1):
    link.append(url+str(i))

for url in link:
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    tag = 'script'
    attributes = {'id':'__NEXT_DATA__'}
    table_soup = soup.find(tag, attributes)
    for item in table_soup:
        x = json.loads(item)
        y = x['props']['pageProps']['cars']
        df2 = pd.DataFrame(y)
        df = pd.concat([df,df2])
```

```python
df['Transmission'] = 'automatic'
df
```
