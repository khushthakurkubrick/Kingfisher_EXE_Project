```python
def scroll_down(self):
    """A method for scrolling the page."""

    # Get scroll height.
    last_height = self.execute_script("return document.body.scrollHeight")

    while True:

        # Scroll down to the bottom.
        self.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load the page.
        time.sleep(3)

        # Calculate new scroll height and compare with last scroll height.
        new_height = self.execute_script("return document.body.scrollHeight")

        if new_height == last_height:

            break

        last_height = new_height
```

```python
#scraping AMT transmission

urls_amt = [r'https://jiji.ng/cars?filter_attr_90_transmission=AMT&filter_attr_119_year_of_manufacture__min=1990&filter_attr_119_year_of_manufacture__max=2011', r'https://jiji.ng/cars?filter_attr_90_transmission=AMT&filter_attr_119_year_of_manufacture__min=2012']

headings = []
prices = []
conditions = []
locations = []
mileages = []
make = []
state = []
city = []
image = []

for car in urls_amt:
    driver = webdriver.Chrome()
    driver.get(car)
    scroll_down(driver)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source)
    full_results = soup.find_all('div',{'class':'b-list-advert__item-wrapper'})
    heading = soup.find_all('div', {'class':'b-advert-title-inner qa-advert-title b-advert-title-inner--h3'})
    price = soup.find_all('div', {'class':'qa-advert-price'})
    condition = soup.find_all('div', {'class':'b-list-advert__item-attr'})
    location = soup.find_all('span', {'class':'b-list-advert__region__text'})
    
    for i in full_results:
        if len(i.find_all('div', {'class':'b-list-advert__item-attr'})) ==3:
            try:
                headings.append(i.find('div', {'class':'b-advert-title-inner'}).get_text().strip())
            except:
                headings.append(None)

            try:
                prices.append(i.find('div', {'class':'qa-advert-price'}).get_text().strip())
            except:
                prices.append(None)      

            try:
                conditions.append(i.find('div', {'class':'b-list-advert__item-attr'}).get_text().strip())
            except:
                conditions.append(None)

            try:
                locations.append(i.find('span', {'class':'b-list-advert__region__text'}).get_text().strip())
            except:
                locations.append(None)

            try:
                state.append(i.find('span', {'class':'b-list-advert__region__text'}).get_text().strip().split(',')[0])
            except:
                state.append(None)

            try:
                city.append(i.find('span', {'class':'b-list-advert__region__text'}).get_text().strip().split(',')[1:][0])
            except:
                city.append(None)

            try:
                mileages.append(i.find_all('div', {'class':'b-list-advert__item-attr'})[2].text.strip()[:10].strip())
            except:
                mileages.append(None)
            
            try:
                image.append(i.find('img')['src'])
            except:
                image.append(None)
    driver.close()

#make from headings
make = []
for g in headings:
        try:
            if g in ['Land Rover', 'Alfa Romeo','Aston Martin',' Great Wall']:
                make.append(g.split()[0:1])
            else:
                make.append(g.split()[0])
        except:
            make.append(None)

years = []
pattern = re.compile('[19|20]{1}[0-9]{3}')
for i in headings:
    try:
        match_result = pattern.search(i)
        years.append(match_result[0])
    except:
        years.append(None)

transmission = ['AMT'] * len(make)
```
